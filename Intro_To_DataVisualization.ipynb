### Line plot
- This plot is useful for visualizing time series data

```python
temperature= np.random.uniform(18,40,size=30)
plt.plot(temperature)
```

```python
# to remove the meta data to show just the plot
temperature= np.random.uniform(18,40,size=30)
plt.plot(temperature)
plt.show()
```

```python
# setting the seed will fix the random numbers
np.random.seed(0)
temperature= np.random.uniform(18,40,size=30)
plt.figure(figsize=(15,5))
plt.plot(temperature)
plt.show()
```

```python
## Plotting multiple line charts
np.random.seed(0)
summer= np.random.uniform(25,40,size=30)
winter= np.random.uniform(7,30,size=30)
rainy= np.random.uniform(15,35,size=30)
plt.figure(figsize=(15,5))
plt.plot(summer)
plt.plot(winter)
plt.plot(rainy)
plt.show()
```

#### Adding attributes

```python
## Plotting multiple line charts
np.random.seed(0)
summer= np.random.uniform(25,40,size=30)
winter= np.random.uniform(7,30,size=30)
rainy= np.random.uniform(15,35,size=30)
plt.figure(figsize=(15,5))
plt.plot(summer, color= 'orange', linestyle=':', lw=4)
plt.plot(winter, color= 'blue', linestyle= '--')
plt.plot(rainy, color= 'green', lw=3)
plt.title('Temperature across seasons')  #Title of the chart
plt.xlabel('Day of the month')  # X axis label
plt.ylabel('Temperature') # Y axis label
plt.legend(['Summer', 'Winter', 'Rainy'],loc=3) # groups with position
plt.show()
```

### Pie chart
- Distribution of different categories

```python
country= ['India', 'China', 'USA', 'Japan', 'Canada']
pop= [145,140,50,20,10]
plt.pie(pop,labels=country)
plt.title('Population per country')
plt.legend(country)
plt.show()
```

### Scatter plot

```python
np.random.seed(0)
height= np.random.uniform(150,185,size=20)
weight= np.random.uniform(150,185, size=20)
for i in range (len(height)):
    weight[i]= height[i]/2+ np.random.randint(-10,10)
plt.title('Height vs Weight')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.scatter(weight,height)
```

```python
np.random.seed(0)
m_height= np.random.uniform(150,185,size=20)
m_weight= np.random.uniform(150,185, size=20)
f_height= np.random.uniform(140,175,size=20)
f_weight= np.random.uniform(150,185, size=20)
for i in range (len(m_height)):
    m_weight[i]= m_height[i]/2+ np.random.randint(-10,10)
for i in range (len(f_height)):
    f_weight[i]= f_height[i]/2+ np.random.randint(-20,10)
plt.title('Height vs Weight')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.scatter(m_weight,m_height)
plt.scatter(f_weight, f_height)
```

### Box and whisker plot

```python
# Salaries of different profiles in an organization
fresher= np.random.uniform(60,100,size=100)
lead= np.random.normal(loc= 120, scale=30, size=50)
manager= np.random.normal(loc= 150, scale=30, size=30)
president= np.random.normal(loc=200, scale=30, size=10)
cxo= np.random.normal(loc= 500, scale=100, size=5)
```

### Histogram

```python
np.random.seed(0)
values= np.random.normal(loc=0, scale=25, size=100)
plt.hist(values, edgecolor='black', bins=15)
plt.show()
## bins= no of bars
```

### Combining plots
### SUBPLOTS

```python
plt.subplot(2,1,1) #(no of rows, no of columns, order)
sns.countplot(x='day', data=tips)

plt.subplot(2,1,2)
sns.heatmap(tips.corr(numeric_only=True),annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) #(no of rows, no of columns, order)
sns.countplot(x='day', data=tips)

plt.subplot(1,2,2)
sns.heatmap(tips.corr(numeric_only=True),annot=True)
plt.show()
```
